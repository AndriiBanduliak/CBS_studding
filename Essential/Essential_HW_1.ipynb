Завдання 2

Створіть клас, який описує відгук до книги. Додайте до класу книги поле – список відгуків. Зробіть так, щоб при виведенні книги на екран за допомогою функції print також виводилися відгуки до неї.

```python
class Review:
    def __init__(self, author, text):
        self.author = author
        self.text = text

    def __str__(self):
        return f"Review by {self.author}: '{self.text}'"

    def __repr__(self):
        return f"Review(author='{self.author}', text='{self.text}')"


class Book:

    def __init__(self, author, title, year, genre):
        self.author = author
        self.title = title
        self.year = year
        self.genre = genre
        self.reviews = []

    def add_review(self, review):
        self.reviews.append(review)
        print(f"A new review has been added to the book '{self.title}'.")

    def __str__(self):
        book_info = (
            f"Book: '{self.title}'\n"
            f"Author: {self.author}\n"
            f"Year: {self.year}\n"
            f"Genre: {self.genre}\n"
        )

        reviews_info = "\n--- Reviews ---\n"
        if self.reviews:
            for review in self.reviews:
                reviews_info += f"- {review}\n"
        else:
            reviews_info += "No reviews yet.\n"

        return book_info + reviews_info

    def __repr__(self):
        return (
            f"Book(author='{self.author}', title='{self.title}', "
            f"year={self.year}, genre='{self.genre}')"
        )


book1 = Book("Erich Maria Remarque", "All Quiet on the Western Front", 1929, "War Novel")
book2 = Book("J.R.R. Tolkien", "The Lord of the Rings", 1954, "Fantasy")
book3 = Book("Franz Kafka", "The Metamorphosis", 1915, "Novella")


review_for_book1 = Review("Alex", "A powerful and heartbreaking story about the realities of war.")
review_for_book2_1 = Review("GandalfFan_1", "An absolute masterpiece of world-building. A must-read!")
review_for_book2_2 = Review("Samwise", "Epic journey with incredible characters.")


book1.add_review(review_for_book1)
book2.add_review(review_for_book2_1)
book2.add_review(review_for_book2_2)


print("="*40)
print(book1)
print("="*40)
print(book2)
print("="*40)

print(book3)

print("--- Demonstration of __repr__ ---")
print([book1, book2, book3])
```

Завдання 4

Створіть клас, який описує автомобіль. Створіть клас автосалону, що містить в собі список автомобілів, доступних для продажу, і функцію продажу заданого автомобіля.

```python

class Car:

    def __init__(self, brand, model, year, price):
        self.brand = brand
        self.model = model
        self.year = year
        self.price = price

    def __str__(self):
        return f"{self.year} {self.brand} {self.model} - Ціна: ${self.price:,.2f}"

    def __repr__(self):
        return f"Car(brand='{self.brand}', model='{self.model}', year={self.year})"

class CarDealership:

    def __init__(self, name):
        self.name = name
        self.cars_for_sale = []
        print(f"Автосалон «{self.name}» відкрито!")

    def add_car(self, car):
        self.cars_for_sale.append(car)
        print(f"Додано новий автомобіль: {car}")

    def list_available_cars(self):
        print(f"\n--- Автомобілі в наявності в «{self.name}» ---")
        if not self.cars_for_sale:
            print("На жаль, автомобілів немає в наявності.")
        else:
            for i, car in enumerate(self.cars_for_sale, 1):
                print(f"{i}. {car}")
        print("-" * (30 + len(self.name)))

    def sell_car(self, car_to_sell):
        if car_to_sell in self.cars_for_sale:
            self.cars_for_sale.remove(car_to_sell)
            print(f"\n Автомобіль {car_to_sell.brand} {car_to_sell.model} продано!")
        else:
            print("\nПомилка: Такого автомобіля немає в наявності.")

    def __len__(self):
        return len(self.cars_for_sale)


car1 = Car("Toyota", "Camry", 2023, 35000)
car2 = Car("Ford", "Mustang", 2022, 55000)
car3 = Car("Honda", "Civic", 2024, 28000)


dealership = CarDealership("SoupAuto")


dealership.add_car(car1)
dealership.add_car(car2)
dealership.add_car(car3)


dealership.list_available_cars()


print(f"\nЗагальна кількість автомобілів в салоні: {len(dealership)}")


dealership.sell_car(car2)


dealership.list_available_cars()
print(f"Кількість автомобілів після продажу: {len(dealership)}")
```
